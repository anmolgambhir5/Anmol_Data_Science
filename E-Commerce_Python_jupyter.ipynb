```python
# 5. Finding the total revenue over all orders placed
def calculate_revenue(dp):
    if(dp>600):
        rev=0.25*dp
    elif(dp>350 and dp<=600):
        rev=0.15*dp
    elif(dp>100 and dp<=350): 
        rev=0.10*dp
    else :
        rev=0.05*dp
    return rev
    
total_revenue=0
for index,row in df.iterrows():
    dp=row[12]
    rev=calculate_revenue(dp)
    total_revenue=total_revenue+rev
print(total_revenue)
    
```

```python
# 6.Calculating Total Brand Revenue and arranging the brands according to the revenue in descending order
total_brand_revenue=0
for index,row in df.iterrows():
    dp=row[12]
    rev=calculate_revenue(dp)
    brand_rev=dp-rev
    total_brand_revenue=total_brand_revenue+brand_rev
    df.at[index, 'brand_revenue'] = brand_rev
print(total_brand_revenue)

df['brand_revenue'].sum()

df.groupby('Brand')['brand_revenue'].sum().sort_values(ascending=False).head(10)
#df.drop(columns=['brand_revenue'],inplace=True)
```
